Cosine similarity metric will be used to estimate similarities between description's.



```python
barrier100=100



from sklearn.metrics.pairwise import cosine_similarity
from numpy import dot
from numpy.linalg import norm
import math
from scipy import spatial

countera=0

final_dict_uni={}
for i in range ( len ( dataframe2)):
  list_100_similar=[]
  for k in range ( len ( dataframe2)):
    countera+=1
    if i==k:
      continue
    else:
      #print ( " Book 1 id = " ,dataframe2['bookId'].iloc[i])

      a = dataframe2['Unigram_vectors'].iloc[i]
      b=dataframe2['Unigram_vectors'].iloc[k]
      #c=dataframe2['description'].iloc[k]

      cos_sim=dot(a, b)/(norm(a)*norm(b))
      if math.isnan(cos_sim):
        cos_sim=0.0
      #print ( cos_sim,"\n\n")

      sort_book_entry=[dataframe2['bookId'].iloc[k], cos_sim ]
      #print ( " Is geometrically close with ", sort_book_entry,"\n\n")


      ###### Inserting in the list which will become the dictionary #####

      #if best 100 list is empty insert the first elemnt you examine
      if len(list_100_similar)==0:
        list_100_similar.append(sort_book_entry)

      else:#if it is not empty instert the element you examine in the correct position

        #SORTINSERT
        for s in range( len(list_100_similar) ):

          # if it is smaller than current index
          if list_100_similar[s][1] >=sort_book_entry[1]:

            indy=s

            # if it is not full yet add everyone
            if len(list_100_similar) <barrier100:
              list_100_similar=list_100_similar[:indy] + [sort_book_entry] + list_100_similar[indy:]

              break


            #if it is Full you have to add in correct position
            else:
              #if it is smaller than the smallest
              #if list_100_similar[0][1] >sort_book_entry[1] :
              if s==0:

                break
              #if it smaller than the current and the current is not the smallest
              # add it here
              else:

                list_100_similar=list_100_similar[1:indy] + [sort_book_entry] + list_100_similar[indy:]
                break


          # if it is bigger than current index
          else :


            #if it is the end of the list
            if s == len(list_100_similar)-1:

              #add it and remove the first one if it is foul
              if s==barrier100-1:

                list_100_similar=list_100_similar[1:s]+ [sort_book_entry]
                break
              # add it at the end if it is not foul
              else :

                list_100_similar=list_100_similar + [sort_book_entry]
                break

            else:

              continue


  list_100_similar.reverse()
  final_dict_uni.update ({ dataframe2['bookId'].iloc[i] : list_100_similar })
  #print (len (list_100_similar), list_100_similar, '\n\n\n')


"""
for i in final_dict_uni:
  print ( " book is : ",i, '\n and the most similar books in increasing order are ', final_dict_uni[i] , '\n\n ')
"""
```
